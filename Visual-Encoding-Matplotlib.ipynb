```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
data = pd.read_csv('Pokemon.csv')
```

# Grouped bar chart

```python
N = 3
# reading attribute values as variable
pokemon_hp = data['HP'][:N]
pokemon_attack = data['Attack'][:N]
pokemon_def = data['Defense'][:N]
pokemon_spatk = data['Sp. Atk'][:N]
pokemon_spdef = data['Sp. Def'][:N]
pokemon_speed = data['Speed'][:N]

ind = np.arange(N) 
width = 0.10       
#plotting each attribute as a separate bar with width increasing in multiples
plt.bar(ind, pokemon_hp, width, label='HP')
plt.bar(ind + width, pokemon_attack, width,label='Attack')
plt.bar(ind + 2*width, pokemon_def, width, label='Defense')
plt.bar(ind + 3*width, pokemon_spatk, width, label='Sp. Atk')
plt.bar(ind + 4*width, pokemon_spdef, width, label='Sp. Def')
plt.bar(ind + 5*width, pokemon_speed, width, label='Speed')


# setting different attributes of the graph
plt.ylabel('Values')
plt.title('Pokemon Attributes')

plt.xticks(ind + 5*width / 2, (data['Name'][:N]),rotation=90)
plt.legend(loc='best')
plt.xlabel('Pokemon')
plt.show()
```

 
 - <b> Encodings I decided not to use, and why </b>: Pie charts for each Pokemon on X axis. Because Angle judgements on area marks are less accurate than length judgements on line marks I decided on using bar charts instead. The wedges vary in width along the radial axis, from narrow near the center to wide near the outside, making the area judgement particularly difficult for pie charts. (VAD page 168) P.S. the task is to visualize the performance of each pokemon and compare them.
 - <b> Advantages and disadvantages </b> : I used a grouped bar chart to represent how different attributes of the three pokemons vary. Using length as channel and bar as mark it is easy to interpret the attributes and compare them across each pokemon. (Marks and Channels Jan 23) However, it is difficult to judge the overall performance of the pokemon on such a grouped bar chart. I need a representation that could show how the pokemons compare with each other overall across attributes.
 

# Stacked bar chart

```python
N = 3
# reading attribute values as variable
pokemon_hp = data['HP'][:N]
pokemon_attack = data['Attack'][:N]
pokemon_def = data['Defense'][:N]
pokemon_spatk = data['Sp. Atk'][:N]
pokemon_spdef = data['Sp. Def'][:N]
pokemon_speed = data['Speed'][:N]

ind = np.arange(N) 
width = 0.50       
# stacking each attribute over one another
plt1 = plt.bar(ind, pokemon_hp, width, label='HP')
plt2 = plt.bar(ind, pokemon_attack, width, bottom = pokemon_hp, label='Attack')
plt3 = plt.bar(ind, pokemon_def, width, bottom = pokemon_attack+pokemon_hp, label='Defense')
plt4 = plt.bar(ind, pokemon_spatk, width, bottom = pokemon_attack+pokemon_hp+pokemon_def, label='Sp. Atk')
plt5 = plt.bar(ind, pokemon_spdef, width, bottom = pokemon_attack+pokemon_hp+pokemon_def+pokemon_spatk, label='Sp. Def')
plt6 = plt.bar(ind, pokemon_speed, width, bottom = pokemon_attack+pokemon_hp+pokemon_def+pokemon_spatk+pokemon_spdef, label='Speed')


# setting graph attributes
plt.ylabel('Values')
plt.title('Pokemon Attributes')

plt.xticks(ind +width/ 20, (data['Name'][:N]),rotation=90)
plt.legend(loc='best')
plt.xlabel('Pokemon')
plt.show()
```

 
 - <b> Encodings I decided not to use, and why </b>: Normalized stacked bar charts. A stacked bar chart uses a composite glyph made of stacking multiple sub-bars of different colors on top of each other; a normalized stacked bar chart stretches each of these bars to the maximum possible length, showing percentages rather than absolute counts. As percentages across different pokemons cannot be compared I decided on using a stacked bar chart without normalization (VAD page 169)P.S. the task is to visualize the performance of each pokemon and compare them.
 - <b> Advantages and disadvantages </b> : I used a stacked bar chart to represent how different attributes of the three pokemons vary. Using length as channel and bar as mark it is easy to interpret the attributes and compare them across each pokemon. (Marks and Channels Jan 23) However, it is difficult to interpret and compare individual attributes across the pokemons as they have varying positions for each attribute with no fixed base line. I need a representation that could show how the pokemons compare with each other at individual attribute level which is where a grouped bar chart comes in handy.
 

- <b> which one of the two visualizations I prefer, and why : </b> I would prefer the stacked bar chart as it encodes an additional attribute which is the overall strength of the pokemon along with the other individual attributes.
